```python
import requests, faker
from pprint import pprint
```

```python
path = "/api/events/"
base_url = "http://localhost:8002"
endpoint = f"{base_url}{path}"
print(endpoint)
```

```python
pages = [
    "/home",
    "/about",
    "/contact",
    "/pricing",
    "/login",
    "/register",
    "/dashboard",
    "/settings",
    "/profile",
    "/logout",
    "/admin",
]
pprint(pages)
```

```python
faker = faker.Faker()
session_ids = [faker.uuid4() for _ in range(20)]
pprint(session_ids)
```

```python
referrers = [
    "https://google.com",
    "https://facebook.com",
    "https://twitter.com",
    "https://linkedin.com",
    "https://instagram.com",
    "https://youtube.com",
    "https://github.com",
    "https://gitlab.com",
    "https://bitbucket.org",
    "https://stackoverflow.com",
    "https://medium.com",
    "https://dev.to",
    ""
    ]
pprint(referrers)
```

```python
from random import choice, randint

events  = 1000
for i in range(events):
    page = faker.random_element(pages)
    user_agent = choice([
                    faker.safari,
                    faker.chrome,
                    faker.firefox,
                    faker.opera,
                    faker.internet_explorer])()
    payload = dict(
        page=page,
        user_agent=user_agent,
        ip_address=faker.ipv4(),
        referrer=faker.random_element(referrers),
        session_id=faker.random_element(session_ids),
        duration= randint(1, 5000),
    )
    response = requests.post(endpoint, json=payload)
    status_code = response.status_code
    print(status_code)
    if response.ok:
        data = response.json()
        pprint(data)
    else:
        print(response.status_code, response.text)

```
