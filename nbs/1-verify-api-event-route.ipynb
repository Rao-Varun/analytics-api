```python
import requests
```

```python
base_url = "http://localhost:8002"
```

```python
pages = [
    "/home",
    "/about",
    "/contact",
    "/pricing",
    "/login",
    "/register",
    "/dashboard",
    "/settings",
    "/profile",
    "/logout",
    "/admin",
]
```

```python
duration = "5 minutes"
limit = 20
```

```python
from pprint import pprint
path = "/api/events"
endpoint = f"{base_url}{path}"
print(endpoint)
response = requests.get(endpoint, params=dict(duration=duration, pages=pages, limit=limit))
print(response.status_code)
print(response.ok)
if response.ok:
    pprint(response.json())
```

```python
path = "/api/events/read_events_by_agents"
endpoint = f"{base_url}{path}"
print(endpoint)
response = requests.get(endpoint, params=dict(duration=duration, limit=limit))
print(response.status_code)
if response.ok:
    pprint(response.json())
else:
    print(response.status_code, response.text)
```
